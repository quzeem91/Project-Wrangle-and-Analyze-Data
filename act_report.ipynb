## Data Wrangling -Act Report 

image from perxel.com

This report details how  the WeRateDogs Twitter data ( the tweet archive of Twitter user **@dog_rates**, also known as **WeRateDogs.**)was wrangled to create interesting and trustworthy analyses and visualizations.The WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. <br>


While the WeRateDogs Twitter Archive Data contains some interesting columns ,Two vital Data were left out(Likes and retweets).<br> .To get the likes and retweets of each tweet_id in the WeRateDogs twitter archive data ,It was required for a twitter Developer Account to be Created(prerequiste,one need to have a twitter account).<br>
    After Creating the Developer Account ,a mail was written to twitter support system requesting for access to their API and how Data will be queried which took a couple of mails before  access was given and approved .<br>
Using the tweet_id in the WeRateDogs Twitter Archive , the json Data of each tweet_id was written to a .txt file ,where  the tweet_id ,likes and retweets were appended to a list (df_lists) ,which was converted to a Dataframe

Then the visual and programmatic Assessment approaches were both taken(Prior to this  ,all Datasets have been loaded into the work environment) were Quality and Tidiness Issues were detected and documented . <br>
The Define,Code and Test approach was employed in the data cleaning  phase of the project and it was iterative .New columns were extracted/calculated such as the (Day,month ,year columns from the timestamp column ) and also few columns that werent pivotal to the analysis were dropped.Then analyses and visualizations were carried out and insights were made.
